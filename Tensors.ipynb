# **Tensors:-**
     Tensors are the data stucture.The work that array done in 1D the ame work tensore does in N-D.

# **0-Dimension Tensor: (scalar)**
           a single number
***Example:-***  after a forward pass it computes a single scalar value indicating the difference between the predicted and actual outputs.
           
# **1-Dimension Tensor: (vector)**
           arrays
           a single number
***Example:-***  in NLP each word in a sentence may be represented as a 1D vector using embeddings.

# **2-Dimension Tensor: (matrices)**
           gray scale images
           
# **3-Dimension Tensor:**
           RGB image
           width x height x channel

# **4-Dimension Tensor:**
           Batches of RGB images
           one video
           batch size x width x height x channel

# **5-Dimension Tensor:**
           video data
           no of videos x no of frames x width x height x channel

# **Why are tensors useful?**


1.   You can easily perform mathematical operations (addition, muliplication, dot product) necessary for neural network operations.
2.   Data like audio,videos,images,text can be easily represented in tensors.

1.   Tensors can run on GPUs and can perfrm many operations parallely and can boost the speed of operations.
