# **The nn Module**
the torch.nn module in python is a core library that provides a wide array of classes and functions designed to help developers build neural network efficiently and effectively. It abstract the complexity of creating and training neural networks by offering pre built layers, loss functions, activation functions, and other utilities enabling you to focus on designing and experimenting with model architecture.

# Key Components of torch.nn
### **Modules(layers)**
**nn.Module:** The base class for all neural networks modules.

**Common Layers:** Includes layers like nn.Linear(fully connected layers), nn.Conv2d(convolutional layers), nn.LSTM (recurrent layers).

### **Activation functions**
functions like nn.Relu, nn.sigmoid, nn.Tanh introduce non-linearities to the model, allowing it to learn complex patterns.

### **Loss Functions**
provide loss functions such as nn.CrossEntropyLoss, nn.MSE loss, and nn.NLLLoss to quantify the difference between the model's prediction and actual target.

### **Container Modules**
**nn.Sequential:** A sequential container to stack layers in order.

### **Regularization and Dropout**
layers linke nn.Dropout and nn.BatchNorm2d help prevent overfitting and improve the model's ability to generalize to new data.

```python
import torch
import torch.nn as nn
```

```python
# create model class
class Model(nn.Module):
  def __init__(self,num_features):
    super().__init__()
    self.network=nn.Sequential(nn.Linear(num_features,3),
    nn.ReLU(),
    nn.Linear(3,1),
    nn.Sigmoid())

  def forward(self,features):
    out=self.network(features)
    return out
```

```python
# create dataset
features=torch.rand(10,5)

# create model
model=Model(features.shape[1])

# call model for forward pass
model(features)
```

# **Torch.optim Module**
torch.optim is a module in pytorch that provides the variety of optimization algorithm used to update the parameters of your model during training. It includes common optimizer like SGD, Adam, RMS prop, and more.

The **model.parameters()** method in pytorch receives an iterator over all the trainable parameters **(weights, biases)** in a model.
